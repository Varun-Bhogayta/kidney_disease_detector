```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pandas as pd 
import numpy as np
```

```python
model = load_model("ckd_ann_model.h5")
```

```python
input = {
    "Age" : 43 ,
    "Creatinine_Level" : 2.00 ,
    "BUN" : 23, 
    "Diabetes" : 1, 
    "Hypertension": 1,
    "GFR" : 62.5,
    "Urine_Output" : 1100
}
43,2.05,22.8,1,1,62.2,1154.0,0,0
```

```python
input_df = pd.DataFrame([input])
input_df
```

```python
import joblib
scaler = joblib.load("scaler.pkl")
input_df_scaled = scaler.transform(input_df)
input_df_scaled
```

```python
if model.predict(input_df_scaled) > 0.5:
    print("patient have CKD.")
else :
    print("patient does not have CKD.")
```
